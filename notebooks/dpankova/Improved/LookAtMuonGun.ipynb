```python
from __future__ import division
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import glob
```

```python
data = np.load('/fastio2/dasha/double_pulse/data_MuonGun_21315.npy', mmap_mode='r')
```

```python
def PlotW2(Name,Label,data,weight,xscale='linear',b =200,r =[0,1], loc =8):
    fig = plt.figure(figsize=(20, 5),facecolor ='w')
    ax = fig.add_subplot(111)
    plt.rcParams.update({'font.size': 16})
    plt.ticklabel_format(axis='both', style='sci', scilimits=(-5,5))
    ax.set_xscale(xscale)
    ax.set_yscale('log')
    ax.set_title(Name, fontsize = 16)
    ax.text(0.01,0.86,'#MuonGun '+str(len(data)), transform=ax.transAxes, color = "black",fontsize=12)  
    ax.set_xlabel(Label, fontsize = 16)                                                              
    ax.set_ylabel("#Events per Year", fontsize = 16)                                                   
    ax.hist(data, bins =b, weights = weight, histtype = 'step',edgecolor ='k', lw =2, range = r, label = "Corsika")  
    
    plt.legend(loc=loc)
```

```python
PlotW2("NET1", "pred", data['preds']['n1'],data['weight_val'])
PlotW2("NET3", "pred", data['preds']['n3'],data['weight_val'])
PlotW2("NET2_1", "pred", data['preds']['n2_1'],data['weight_val'])
PlotW2("NET2_2", "pred", data['preds']['n2_2'],data['weight_val'])
PlotW2("NET2_3", "pred", data['preds']['n2_3'],data['weight_val'])
```
