```python
from __future__ import division
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
diffs = np.load('/home/dup193/work/double_pulse/ATDW_diffs.npy')
print(diffs[0:10])
```

```python
fig = plt.figure(dpi=100, figsize=(20,8))
fig.suptitle('Difference between ATWDs staring times in bins')

#ax=fig.add_subplot(121)
#bins = np.logspace(3, 8, 100)

plt.hist(diffs[:,0], bins=100, histtype='step', color = 'r', label = "ATWDa" )
plt.hist(diffs[:,1], bins=100, histtype='step', color = 'b', label = "ATWDb" )
plt.hist(abs(diffs[:,1]-diffs[:,0]), bins=100, histtype='step', color = 'g', label = "ATWDb-ATWDa" )

plt.grid(True)
#plt.xscale('log')
plt.ylabel('N', size=12)
plt.xlabel('Bin Value', size=12)
plt.legend()
```
