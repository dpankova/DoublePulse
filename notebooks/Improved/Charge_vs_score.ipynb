```python
from __future__ import division
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_ec = np.load('/fastio2/dasha/double_pulse/data_NuECC.npy', mmap_mode = 'r')[:100000]
data_mc = np.load('/fastio2/dasha/double_pulse/data_NuMuCC.npy', mmap_mode = 'r')[:100000]
data_tc = np.load('/fastio2/dasha/double_pulse/data_NuTauCC.npy', mmap_mode = 'r')[:100000]
data_en = np.load('/fastio2/dasha/double_pulse/data_NuENC.npy', mmap_mode = 'r')[:100000]
data_mn = np.load('/fastio2/dasha/double_pulse/data_NuMuNC.npy', mmap_mode = 'r')[:100000]
data_tn = np.load('/fastio2/dasha/double_pulse/data_NuTauNC.npy', mmap_mode = 'r')[:100000]
data_eg = np.load('/fastio2/dasha/double_pulse/data_NuEGR.npy', mmap_mode = 'r')[:100000]


preds_tc = np.load('/fastio2/dasha/double_pulse/preds_NuTauCC_400_Q2000_NO.npy')[:100000]
preds_mc = np.load('/fastio2/dasha/double_pulse/preds_NuMuCC_400_Q2000_NO.npy')[:100000]
preds_ec = np.load('/fastio2/dasha/double_pulse/preds_NuECC_400_Q2000_NO.npy')[:100000]
preds_tn = np.load('/fastio2/dasha/double_pulse/preds_NuTauNC_400_Q2000_NO.npy')[:100000]
preds_mn = np.load('/fastio2/dasha/double_pulse/preds_NuMuNC_400_Q2000_NO.npy')[:100000]
preds_en = np.load('/fastio2/dasha/double_pulse/preds_NuENC_400_Q2000_NO.npy')[:100000]
preds_eg = np.load('/fastio2/dasha/double_pulse/preds_NuEGR_400_Q2000_NO.npy')[:100000]
```

```python
fig = plt.figure(figsize=(20, 10),facecolor ='w')
ax = fig.add_subplot(111)
plt.rcParams.update({'font.size': 16})
plt.ticklabel_format(axis='both', style='sci', scilimits=(-5,5))
ax.set_xscale('log')
ax.set_yscale('log')
qt= 5000000
bins = np.logspace(np.log10(100),np.log10(qt), 50)

ax.set_title("Total Charge", fontsize = 16)
ax.text(0.01,0.95,'#NuTauCC '+str(len(data_tc)), transform=ax.transAxes, color = "black",fontsize=12)
ax.text(0.01,0.92,'#NuMuCC '+str(len(data_mc)), transform=ax.transAxes, color = "black",fontsize=12)
ax.text(0.01,0.89,'#NuECC '+str(len(data_ec)), transform=ax.transAxes, color = "black",fontsize=12)
ax.text(0.01,0.86,'#NuTauNC '+str(len(data_tn)), transform=ax.transAxes, color = "black",fontsize=12)
ax.text(0.01,0.83,'#NuMuNC '+str(len(data_mn)), transform=ax.transAxes, color = "black",fontsize=12)
ax.text(0.01,0.80,'#NuENC '+str(len(data_en)), transform=ax.transAxes, color = "black",fontsize=12)
ax.text(0.01,0.77,'#NuEGR '+str(len(data_eg)), transform=ax.transAxes, color = "black",fontsize=12)

ax.set_xlabel('Charge', fontsize = 16)                                                              
ax.set_ylabel("#Events", fontsize = 16)                                                   
ax.hist(data_tc['q_tot'], bins =bins, histtype = 'step',edgecolor ='r', lw =2, label = "NuTauCC")  
ax.hist(data_mc['q_tot'], bins =bins, histtype = 'step',edgecolor ='b', lw =2, label = "NuMuCC")  
ax.hist(data_ec['q_tot'], bins =bins, histtype = 'step',edgecolor ='g', lw =2, label = "NuECC")  
ax.hist(data_tn['q_tot'], bins =bins, histtype = 'step',edgecolor ='r', lw =2, ls ="--", label = "NuTauNC")  
ax.hist(data_mn['q_tot'], bins =bins, histtype = 'step',edgecolor ='b', lw =2, ls ="--", label = "NuMuNC")  
ax.hist(data_en['q_tot'], bins =bins, histtype = 'step',edgecolor ='g', lw =2, ls ="--", label = "NuENC")  
ax.hist(data_eg['q_tot'], bins =bins, histtype = 'step',edgecolor ='g', lw =2, ls =":", label = "NuEGR")  

plt.legend()
plt.show()
```
