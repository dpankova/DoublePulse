```python
LLH = -0.1
Qt,Qst0,Qst1,Qst2 = 1000,2000,10,10
N1 = 0.99
N2 = 0.9
N3 = 0.8
secs_per_year = 31536000

INDEX_0 = 2.88
PHI_0 = 2.1467
INDEX_1 = 2.50
PHI_1 = 2.23
INDEX_2 = 2.37
PHI_2 = 1.36
```

```python
def check(data,weight_key = 'weight_val'):
    data = data[np.isfinite(data['logan_veto']['SPE_rlogl']) & np.isfinite(data['logan_veto']['Cascade_rlogl'])]
    masks = (data['preds']['n1'] >= N1) & (data['preds']['n2_3'] >= N2) & (data['preds']['n3'] >= N3)
    maskl = data['logan_veto']['SPE_rlogl']-data['logan_veto']['Cascade_rlogl'] >= LLH
    maskq = (data['qst']['q'][:,2] >=Qst2) & (data['qst']['q'][:,1] >=Qst1) & (data['qst']['q'][:,0] >=Qst0)
    mask = masks & maskl & maskq
    rate = np.sum(data[weight_key][mask]*secs_per_year) 
    rate_err = np.sqrt(np.sum(np.square(data[weight_key][mask]*secs_per_year)))
    print("Events: {0:d}, rate: {1:.3f}+/-{2:.3f}".format(data.shape[0],rate,rate_err))
```

```python
def Pltvals(d,e,PHI,INDEX):
    fig = plt.figure( figsize=(20, 5), facecolor ='w')
    plt.errorbar(range(len(d)), d.values(), yerr= e.values(),capsize = 2)
    l = list(d.values())
    plt.ylim(np.min(l)-0.05,np.max(l)+0.05)
    plt.text(0,np.max(l)+0.03 ,'Phi = {0:.2f}, Gamma = -{1:.2f}'.format(PHI,INDEX), fontsize =16)
    plt.hlines(y=d.values(), xmin=0, xmax=range(len(d)), color='k', ls = '--')
    plt.xticks(range(len(d)), list(d.keys()))
    plt.show()
```

```python
def Plthist(d,PHI,INDEX,b):
    fig = plt.figure( figsize=(20, 5), facecolor ='w')
    l = list(d.values())
    rmin = np.min(l)-0.1
    rmax = np.max(l)+0.1
    n, bins, patches = plt.hist(d.values(), bins =b, range = [rmin,rmax],histtype = 'step')
    (mu, sigma) = norm.fit(l, floc = l[0])
    y = norm.pdf(np.linspace(rmin,rmax,100), mu, sigma)
    plt.text(rmin,np.max(n)+1.5,'Phi = {0:.2f}, Gamma = -{1:.2f}'.format(PHI,INDEX), fontsize =16)
    plt.text(rmin,np.max(n)+1,'Mean = {0:.2f}, STD = {1:.2f}'.format(mu,sigma), fontsize =16)
    l = plt.plot(np.linspace(rmin,rmax,100), y, 'r--', linewidth=2)
    plt.ylim(0,np.max(y)+1)
    plt.show()
```

```python
Plthist(d1,PHI_0,INDEX_0,9)
Plthist(d2,PHI_1,INDEX_1,9)
Plthist(d3,PHI_2,INDEX_2,9)
```
