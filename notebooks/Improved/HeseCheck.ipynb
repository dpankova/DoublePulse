```python
ID_T = np.dtype([('run_id', '<u4'), ('sub_run_id', '<u4'), ('event_id', '<u4'), ('sub_event_id', '<u4')])
ARY_T = np.dtype([("id", ID_T), ("hese_qtot", np.float32), ("hese_llhratio", np.float32)])

def GetHESE(data,Data):
    set_data = {tuple(x[0]) for x in data['id']}
    hese = np.zeros(0,dtype=ARY_T)
    for m,i in enumerate(Data):        
        print("\rPercent = "+str(m/Data.shape[0]*100)+" "+str(m)+" of "+str(Data.shape[0]), end="")
        if tuple(i['id'][0]) in set_data:
            temp = np.zeros(1,dtype = ARY_T)
            temp[["id","hese_qtot","hese_llhratio"]] = (i['id'][0], i['hese_qtot'][0], i['hese_llhratio'][0])
            hese = np.append(hese, temp)
           # print("aaaa",hese)
    idxh = np.argsort(hese, order ='id')
    return idxh, hese
```

```python
D_T = np.dtype([('run_id', '<u4'), ('sub_run_id', '<u4'), ('event_id', '<u4'), ('sub_event_id', '<u4')])
ARX_T = np.dtype([("id", ID_T), ("x", np.float32)])
def GetDataIDX(data):
    data_2 = np.zeros(0,dtype=ARX_T)
    for i in data:
        temp = np.zeros(1,dtype = ARX_T)
        temp[["id","x"]] = (i['id'][0], 0)
        data_2 = np.append(data_2, temp)
    idxd = np.argsort(data_2, order ='id')
    return idxd
```

```python
def SortHESE(data,Data):
    idxh, hese2 = GetHESE(data,Data)
    idxd = GetDataIDX(data)
    idx = zip(idxh,idxd)
    hese = np.zeros(len(hese2),dtype=ARY_T)
    for ih, idt in idx:
        hese[idt] = hese2[ih]
    return hese
```
