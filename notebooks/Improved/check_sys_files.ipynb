```python
def check(file_name):
    data = np.load(file_name, mmap_mode="r")['arr_0']
    fig = plt.figure(figsize=(15, 5))
    ax1 = fig.add_subplot(131)
    ax1.imshow(data[0]['image'][0,:,:,0], interpolation='nearest', aspect='auto', cmap= 'gray')
    ax2 = fig.add_subplot(132)
    ax2.imshow(data[0]['image'][0,:,:,1], interpolation='nearest', aspect='auto', cmap= 'gray')
    ax3 = fig.add_subplot(133)
    ax3.imshow(data[0]['image'][0,:,:,2], interpolation='nearest', aspect='auto', cmap= 'gray')
```
