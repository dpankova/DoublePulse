```python
plot_data = []
plot_data0 = []
plot_data2 = []
plot_data3 = []
plot_data4 = []
plot_data5 = []
plot_data6 = []
plot_data7 = []

for c in cuts:
    tp = len([i for i in preds_t if i[0]>=c])/preds_t.shape[0]*100
    fp = len([i for i in preds_e if i[0]>=c])/preds_e.shape[0]*100
    tp2 = len([i for i in preds2_t if i[0]>=c])/preds2_t.shape[0]*100
    fp2 = len([i for i in preds2_e if i[0]>=c])/preds2_e.shape[0]*100
    tp3 = len([i for i in preds3_t if i[0]>=c])/preds3_t.shape[0]*100
    fp3 = len([i for i in preds3_e if i[0]>=c])/preds3_e.shape[0]*100
    tp4 = len([i for i in preds4_t if i[0]>=c])/preds4_t.shape[0]*100
    fp4 = len([i for i in preds4_e if i[0]>=c])/preds4_e.shape[0]*100
    tp5 = len([i for i in preds5_t if i[0]>=c])/preds5_t.shape[0]*100
    fp5 = len([i for i in preds5_e if i[0]>=c])/preds5_e.shape[0]*100
    tp6 = len([i for i in preds6_t if i[0]>=c])/preds6_t.shape[0]*100
    fp6 = len([i for i in preds6_e if i[0]>=c])/preds6_e.shape[0]*100
    tp7 = len([i for i in preds7_t if i[0]>=c])/preds7_t.shape[0]*100
    fp7 = len([i for i in preds7_e.reshape(preds7_e.shape[0],2) if i[0]>=c])/preds7_e.shape[0]*100
    tp0 = len([[i[1],i[0]] for i in preds0_t if i[1]>=c])/preds0_t.shape[0]*100
    fp0 = len([[i[1],i[0]] for i in preds0_e if i[1]>=c])/preds0_e.shape[0]*100
    
    plot_data.append([c,tp,fp])
    plot_data2.append([c,tp2,fp2])
    plot_data3.append([c,tp3,fp3])
    plot_data4.append([c,tp4,fp4])
    plot_data5.append([c,tp5,fp5])
    plot_data6.append([c,tp6,fp6])
    plot_data7.append([c,tp7,fp7])

    plot_data0.append([c,tp0,fp0])

plot_data = np.array(plot_data)
plot_data2 = np.array(plot_data2)
plot_data3 = np.array(plot_data3)
plot_data4 = np.array(plot_data4)
plot_data5 = np.array(plot_data5)
plot_data6 = np.array(plot_data6)
plot_data7 = np.array(plot_data7)

plot_data0 = np.array(plot_data0)
```

```python
c1 = "#e41a1c"
c2 = "#377eb8"
c3 = "#4daf4a"
c4 = "#984ea3"
c5 = "#ff7f00"
c6 = "gold"#"#ffff33"

def Plot(arr1, arr2, arr3, arr4, arr5, arr6, arr7, arr0, name, bs =100):
    fig = plt.figure(figsize=(10, 10),facecolor = "w")
    plt.rcParams.update({'font.size': 14})
    ax = fig.add_subplot(111)
    ax.set_title(name, fontsize = 16)
    ax.set_ylabel("True positive, NuTau as NuTau", fontsize = 16)                                                                                                                           
    ax.set_xlabel("False positive, NuE as NuTau", fontsize = 16)                                                   
    ax.set_xlim(0,100)
    ax.set_ylim(0,100)
    plt.plot(arr1[:,2], arr1[:,1], label ="Norm per set, 200k", lw =2, c = c1)
    plt.plot(arr2[:,2], arr2[:,1], label ="Norm per image, 200k", lw =2, c = c2)
    plt.plot(arr3[:,2], arr3[:,1], label ="Norm per set, 600k", lw =2, c = c3)
    plt.plot(arr4[:,2], arr4[:,1], label ="Norm per set, 200k, Qst=2000", lw =2, c = c4)
    plt.plot(arr5[:,2], arr5[:,1], label ="Norm per set, 400k, Qst=2000", lw =2, c = c5)
    #plt.plot(arr6[:,2], arr6[:,1], label ="Current")
    plt.plot(arr7[:,2], arr7[:,1], label ="VGG16", lw =2, c = c6)
    plt.plot(arr0[:,2], arr0[:,1], label ="Benchmark", ls ='--', c='k', lw =2)
    plt.xticks(np.arange(0, 101, 5))
    plt.yticks(np.arange(0, 101, 5))
    
    plt.xlim([0,10])
    plt.ylim([0,20])
    plt.grid()
    plt.legend()
```
