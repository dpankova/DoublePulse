```python
from __future__ import division
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
keys = []
for i in range(1,14):
    name = "/home/dup193/work/double_pulse/data/key_count/GetKeys_{0}_keys.pkl".format(i)
    keys_temp = pickle.load(open(name, "rb"))
    keys = keys + keys_temp
keys = np.array(keys)
```

```python
tot = len(keys)
print(tot)
print(keys[0:10])
```

```python
num_null_split = 0
for e in keys:
    if e['NullSplit'] == False:
        num_null_split = num_null_split +1
print(num_null_split/tot)
```

```python
no_null = []
for e in keys:
    if e['NullSplit'] == False:
        no_null.append(e)
tot = len(no_null)
print(tot)
print(no_null[:10])
```

```python
no_raw_data = 0
no_pulses = 0
no_weights = 0
no_mcTree = 0
no_qtot = 0
no_header = 0
no_all = 0

for e in no_null:
    if e['InIceRawData'] == False:
        no_raw_data += 1
    if e['SplitInIcePulses'] == False:
        no_pulses += 1
    if e['I3MCWeightDict'] == False:
        no_weights += 1
    if e['I3MCTree'] == False:
        no_mcTree += 1
    if e['Homogenized_QTot'] == False:
        no_qtot += 1
    if e['I3EventHeader'] == False:  
        no_header += 1
    if (e['InIceRawData'] == False) or (e['SplitInIcePulses'] == False) or (e['I3MCWeightDict'] == False) or (e['I3MCTree'] == False) or (e['Homogenized_QTot'] == False) or (e['I3EventHeader'] == False):
        no_all += 1
```

```python
print(no_raw_data, no_pulses, no_weights, no_mcTree, no_qtot, no_header, no_all)
```

```python
fig = plt.figure(figsize=(15,5),facecolor='w')
objects = list(no_null[0].keys())
objects.remove('NullSplit')
objects.append('Any')
print(objects)
y_pos = np.arange(len(objects))
perc = [no_raw_data, no_pulses, no_weights, no_mcTree, no_qtot, no_header, no_all]
perc = [i/tot*100 for i in perc]
plt.bar(y_pos, perc, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.yscale('log')
plt.ylabel('Percent lost')
plt.title('Key name')

plt.show()
```

```python
def Plot(arr1, arr2, name, bs =100, r = [1000,1000000], lab = "Energy, GeV", scale = "linear"):
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111)
   # plt.ticklabel_format(axis='both', style='sci', scilimits=(-5,5))
    weights1 = np.ones_like(arr1)/float(len(arr1))
    weights2 = np.ones_like(arr2)/float(len(arr2))
    
    #ax.set_xscale(scale)
    #ax.set_yscale('log')
    ax.set_title(name, fontsize = 14)
    ax.text(0.5,0.95,'#Events '+str(len(arr1)), transform=ax.transAxes, color = "black",fontsize=12)
    ax.text(0.5,0.90,'#Events '+str(len(arr2)), transform=ax.transAxes, color = "black",fontsize=12)
    
    ax.set_xlabel(lab, fontsize = 14)                                                                                                                           
    #ax.set_ylabel("FractionEvents", fontsize = 14)                                                   
    ax.hist(arr1, bins =bs, range = r, histtype = 'step',edgecolor ='r', fill= False)  
    ax.hist(arr2, bins =bs, range = r, histtype = 'step',edgecolor ='b', fill= False)  
```

```python
Plot(m_no_rd_e, m_rd_e, 'NuTauCC', r= [0, 200000], scale = 'linear')
```
