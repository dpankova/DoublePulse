```python
from __future__ import division
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
c =0.299792458
n =1.3195
v=c/n 

x= np.linspace(-2,2,5)*125
y= np.linspace(-2,2,5)*17
dx,dy = np.meshgrid(x, y)
plt.plot(dx, dy, marker='.', color='k', linestyle='none')
plt.show()
print(len(dx))
```

```python
b1 = [-50,0]
b2 = [50,0]
distb = ((b1[0]-b2[0])**2+(b1[1]-b2[1])**2)**0.5
tb = distb/v 

plt.plot(dx, dy, marker='.', color='k', linestyle='none')
plt.plot(b1[0], b1[1], marker='.', color='r', linestyle='none')
plt.plot(b2[0], b2[1], marker='.', color='g', linestyle='none')

plt.show()
```

```python
dom_list = []
for i in range(0,len(dx)):
    for j in range(0,len(dx[0])):
        x = dx[i][j]
        y = dy[i][j]
        dist1 = ((b1[0]-x)**2+(b1[1]-y)**2)**0.5
        dist2 = ((b2[0]-x)**2+(b2[1]-y)**2)**0.5
        t1 = dist1/v
        t2 = dist2/v+tb
        dom_list.append([x,y,t1,t2])
print(dom_list)
temp = sorted(dom_list, key=lambda x: x[1], reverse = True)
print(temp)
```

```python
max_h = 1
def time_plots(list):
    fig = plt.figure(figsize=(10,10))
    fig.subplots_adjust(hspace = .5, wspace=.1)
    for i in list:
        ax = fig.add_subplot(5, 5, list.index(i)+1)
        ax.axvline(i[2], color ='r')
        ax.axvline(i[3], color ='g')
        ax.text(0,max_h*1.2,"({0},{1})".format(i[0],i[1]), color="k",fontsize = 16 )     
        ax.set_ylim(0,max_h*1.4)
        ax.set_xlim(0,2000)
    fig.tight_layout()
    plt.show()
    plt.close('all')
```

```python
time_plots(temp)
```
